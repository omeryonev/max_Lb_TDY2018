**E = Yap elii elastisite modl, E = 200000 N/mm2**

**F_y = Yap eliinin karakteristik akma gerilmesi**

```python
%%render
E = (200000*MPa)
i_y = (30*mm)
F_y = (235*MPa)
suneklik_duzeyi = "sinirli"
```
