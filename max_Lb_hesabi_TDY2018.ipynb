**L_b = Basınç başlığının yanal doğrultuda mesnetlendiği veya enkesitin çarpılmaya karşı mesnetlendiği noktalar arasındaki uzaklık**

**i_y = Kiriş enkesitinin zayıf eksenine göre atalet yarıçapı**

**E = Yapı çeliği elastisite modülü, E = 200000 N/mm2**

**F_y = Yapı çeliğinin karakteristik akma gerilmesi**

## Inputs

```python
%%render
E = (200000*MPa)
i_y = (30*mm)
F_y = (235*MPa)
suneklik_duzeyi = "sinirli"
```

## Outputs

```python
%%render
if suneklik_duzeyi == "yuksek": L_b = 0.086 * i_y * E / F_y 
elif suneklik_duzeyi == "sinirli": L_b = 0.17 * i_y * E / F_y 
```
